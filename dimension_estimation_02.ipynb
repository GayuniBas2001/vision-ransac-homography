2. Fig. 1 shows an image ofa pair of earrings. Assume that a camera is mounted with the
optical axis perpendicular to the imaging plane (where the flat earrings are kept). If the
focal length of the camera is 8 mm, the pixel size is 2.2 µm×2.2 µm, and the distance from
the lens to the imaging plane is 720 mm, what are the sizes of the earrings?

Because the physical dimensions are initially unknown, this script uses OpenCV to extract the pixel width and height of the earrings from the image. Using the principles of the pinhole camera model, these pixel measurements are then converted into real-world metric sizes (millimeters).

Note : I have cropped the image from top and bottom to allow the earrings to be visible within the descktop screen without being scaled. This does not affect the pixel height and width of the earrings.

Extract the pixel dimensions of the earring

```python
import cv2

# Load the image of the earrings
image_path = "earrings.jpg"
img = cv2.imread(image_path)

if img is None:
    print("Error: Could not load image.")
else:
    # Let the user draw a bounding box to measure the earring
    print("Draw a box around one earring and press SPACE or ENTER.")
    roi = cv2.selectROI("Select Earring", img, fromCenter=False, showCrosshair=True)
    cv2.destroyAllWindows()

    # The ROI returns: (x, y, width, height)
    pixel_width = roi[2]
    pixel_height = roi[3]

    print(f"Earring Width in pixels: {pixel_width}")
    print(f"Earring Height in pixels: {pixel_height}")
```
